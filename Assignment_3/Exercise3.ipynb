```python
def naturalNums_Sum(num):
  sum = 0
  for i in range(1,num+1):
    sum+=i
  return sum
num = int(input("Enter a number: "))
print("The sum of first",num,"natural numbers is:")
naturalNums_Sum(num)
```

```python
def reverse_Num(num):
  n = num
  rev_Num = 0
  while(num>0):
      lastDig = num%10
      rev_Num = rev_Num*10 +lastDig
      num//=10
  print("The reverse of",n,"is:",rev_Num)
num = int(input("Enter a number: "))
reverse_Num(num)
```

```python
def countDigits(num):
  count = 0

  while (num>0):
    lastDig = num%10
    count = count + 1
    num//=10
  return count
num = int(input("Enter a number: "))
print("The number of digits in",num,"is:",countDigits(num))
```

```python
def isPallindrome(num):
  res =0
  n = num
  while (num>0):
    lastDig = num%10
    res = res*10 + lastDig
    num//=10
  if (n==res):
    print("It is a Pallindrome.")
  else:
        print("It is NOT a Pallindrome.")
num = int(input("Enter a number: "))
isPallindrome(num)
```

```python
def fibonacci(num):
  num1 = 0
  num2 = 1
  print(num1,'\n',num2)
  for i in range(3,num+1):
    num3 = num1 + num2
    print(num3)
    num1 = num2
    num2 = num3
num = int(input("Enter a number: "))
fibonacci(num)
```

```python
def add(num1,num2):
  print(num1+num2)

def subtract(num1,num2):
    print(num1-num2)

def div(num1,num2):
    print(num1/num2)

def prod(num1,num2):
  print(num1*num2)

num1 = int(input("Enter number 1: "))
num2 = int(input("Enter number 2: "))
op = input("Enter operation to be performed(+,-,*,/): ")
if (op=='+'):
  add(num1,num2)
elif (op=='-'):
  subtract(num1,num2)
elif (op=='*'):
  prod(num1,num2)
else:
  div(num1,num2)
```
